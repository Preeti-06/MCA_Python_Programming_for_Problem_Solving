```python
import csv
dept = {}
with open("employees.csv") as f:
    reader = csv.reader(f)
    for row in reader:
        d = row[1]
        sal = int(row[2])
        dept.setdefault(d, []).append(sal)
print("Average salary by department:")
for d in dept:
    print(d, ":", sum(dept[d]) / len(dept[d]))
```
