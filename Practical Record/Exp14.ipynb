```python
%%writefile shapes/circle.py
import math

def area(radius):
    return math.pi * radius * radius

def perimeter(radius):
    return 2 * math.pi * radius
```

```python
%%writefile shapes/rectangle.py
def area(length, breadth):
    return length * breadth

def perimeter(length, breadth):
    return 2 * (length + breadth)
```

```python
from shapes.circle import area as circle_area, perimeter as circle_perimeter
from shapes.rectangle import area as rect_area, perimeter as rect_perimeter

print("Circle Area:", circle_area(5))
print("Circle Perimeter:", circle_perimeter(5))

print("Rectangle Area:", rect_area(4, 6))
print("Rectangle Perimeter:", rect_perimeter(4, 6))
```
